#Exercise 2

Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0
    
[PEW] E3.2

````python
hours = float (input ('Enter hours worked:'))
rate = float (input ('Enter hourly rate:'))

if hours>40 :
  pay = 40*rate+((hours-40)*(1.5*rate))
  print('pay:',pay)

else:
  pay=hours*rate
  print('pay:',pay)
````

#Challenge

Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

    Score           Grade
    [ 97, 100]      1.0
    [ 93,  97)		1.25
    [ 89,  93)		1.5
    [ 85,  89)		1.75
    [ 81,  85)		2.0
    [ 77,  81)		2.25
    [ 74,  77)		2.5
    [ 0,   74)		3.0

Adapted from [PEW] E3.3
<br>

````python
score=int(input ('Enter your score:'))

if 97<=100:
  print ('Your grade is : 1.0')
elif 93<=score <=97:
  print ('Your grade is: 1.25')
elif 89<= score <= 93:
  print ('Your grade is: 1.5')
elif 85<= score <=89:
  print ('Your grade is: 1.75')

elif 81<= score <=85:
  print ('Your grade is: 2.0')
elif 77<= score <= 81:
  print ('Your grade is: 2.25')
elif 74<= score <=77:
  print ('Your grade is: 2.5')
elif 0<= score <= 74:
  print ('Your grade is: 3.0')
````
